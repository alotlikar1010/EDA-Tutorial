```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df =pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
country_name =final_df.Country.value_counts().index
country_value =final_df.Country.value_counts().values
```

```python
#pie chart top 3 country
plt.pie(country_value[:3], labels=country_name[:3],autopct="%1.2f%%")
```

```python
#observation: zomato max record/transaction are from india after usa and uk
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
##observation
'''
1. when rating is betn 4.5 to 4.9 ->> excellent
2. when rating is betn 4.0 to 3.4  -->> very good 
3 when rating is betn 3.5 to 3.9  ->> good
4  when rating is betn 3.0 to 3.4  ->> average
5. when rating is betn 2.5 to 2.9  ->> average
6 .when rating is betn 2.0 to 2.4  ->> poor



'''
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
'''
Observation:
Not Rated count is very high
Maximum number of rating are between 2.5 to 3.4
'''
```

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
